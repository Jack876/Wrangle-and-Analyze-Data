### Quality issues

1.Data completeness

There are missing values in some columns in tweet_archive table. For example, in_reply_to_status_id and in_reply_to_user_id have only 78 non-null values. 
retweeted_status_id, retweeted_status_user_id and retweeted_status_timestamp have 181 non-null values. Some values are missing in columns like "doggo","floofer" and "name". These missing data is represented by None instead of NaN. Some original ratings have no image data.

2.Dog name validity issue

In twitter_archive table,the last few rows show some dog names are 'None', 'a', or 'an.'These are obviously wrong.

3.Data type issue

In twitter_archive table, timestamp is an object and retweeted_status_timestamp is also an object (the other retweeted statuses are floats).rating_numerator and rating_denominator data types are int64 where some numbers are float type. Timestamp data type is not right too.

4.Rating data consistency issue

There are three columns including information about rating: text, rating_numerators and rating_denominators which are sometimes not consistent. It confuses data users. 
For instance, when tweets id = 682962037429899265, the rating numerator and rating denominator in the table are respectively 9 and 11. But after checking the text, it is: After so many requests, this is Bretagne. She was the last surviving 9/11 search dog, and our second ever 14/10. RIP. Obviously, these two values should actually be 14/10 instead of 9/11.
  

5.Rating number validity 

There are some strange outliers in columns. From coding(twitter_archive.rating_denominator.value_counts()),there are 23 rating_denominators that are not equal to 10. Similarly, there are rating_numerators that are even bigger than 50. It is hard to compare the rating numbers directly based on rating numerators or rating denominators.  

For instance: the highest rating numerator is 1776 when tweet id is 749981277374128128. This number is too good to be true. So I use its expanded urls to check back its original tweets. The text is This is Atticus. He's quite simply America af. 1776/10. Here is the picture:
On July 2, 1776, the Second Continental Congress, still meeting in Philadelphia, voted unanimously to declare the independence as the "United States of America". Two days later, on July 4, Congress adopted the Declaration of Independence. (From wikipedia). Obviously, 1776 is not the real rating numerator.




Another example, when tweets id = 716439118184652801, the rating numerator and rating denominator in the table are respectively 50 and 50. But when checking the text, it actually is: This is Bluebert. He just saw that both #FinalFur match ups are split 50/50. Amazed af. 11/10. Obviously, these two original rating numbers were wrote down wrongly, they should be 11/10.
 

6.Erroneous data

pupper, puppo, floofer and doggo column: There are some IDs with more than one dog stage information (two dog stages are rated).

7.Duplicated values

From coding (sum(image_prediction.jpg_url.duplicated())),we know there are some duplicated values in this column.

8.Data types are not consistent

In tweet_json table, tweet_id is object type. This is not consistent with data type in other 2 tables(int64 data type),which is not suitable for further data analysis.

9.Table size do not match

There are 2356 rows in twitter_archive, while 2075 rows in images_prediction.The size of the table do not match.

10.Timestamp accuracy

Timestamp column is object instead data time. The values in this column include year, month, day, time and number like "+0000". These numbers are mixed together. "+0000" does not make sense.

11.Dog breeds prediction consistency

The dog breeds prediction information can be found in columns p1, p1_dog, p2, p2_dog, p3 and p3_dog in image_prediction table. These value are not consistent and convenient for further analysis. 


### Tidiness Issues

1 Columns issue.

One of the principle is each variable forms a column. In twitter_archive table, the last four columns all relate to the stage of dogs (dogoo, floofer, pupper, puppo). Similarly, the rating numbers are in two columns (rating_numerator and rating_denominator). This is not convenient and a little confusing. For example, most rating_numerator and rating_denominator are around 10, but some numbers are like 143/130, 121/110, or 204/170 etc.

2 Untidy data.

In twitter_archive table, column source data have <a and /a> tag in HTML format surrounding the text.These are not common internet address for ordinary users.

3 Uppercase and lowercase mixture

In images prediction table, there is no consistency in p1, p2 and p3 columns. Sometimes the dog breed predictions are lowercase, sometimes they are uppercase.

4 Many tables VS one table

Tidiness requires each type of observational unit forms a table. These three tables can be merged into one master dataset to improve data tidiness.

### 9.  Dog breeds prediction

```python
# I get the dog breed prediction data from p1, p2 or p3

dog_breed = []

def image(image_prediction_clean):
    if image_prediction_clean['p1_dog'] == True:
        dog_breed.append(image_prediction_clean['p1'])
        
    elif image_prediction_clean['p2_dog'] == True:
        dog_breed.append(image_prediction_clean['p2'])
       
    elif image_prediction_clean['p3_dog'] == True:
        dog_breed.append(image_prediction_clean['p3'])
       
    else:
        dog_breed.append('False')
        
image_prediction_clean.apply(image, axis=1)

image_prediction_clean['dog_breed'] = dog_breed
```

```python
#drop rows that has false prediction
image_prediction_clean = image_prediction_clean[image_prediction_clean['dog_breed'] != 'False']

image_prediction_clean.info()
```

### 10.  Merge these into one master dataframe

```python
#Firstly, merge twitter_archive_clean and image_prediction_clean
df_twitter_transition = pd.merge(twitter_archive_clean, 
                      image_prediction_clean, 
                      how = 'left', on = ['tweet_id'])

#keep rows that have picture (jpg_url)
df_twitter_transition = df_twitter_transition[df_twitter_transition['jpg_url'].notnull()]
```

```python
#Secondly, merge df_twitter1 and tweet_json_clean
df_twitter = pd.merge(df_twitter_transition, tweet_json_clean, 
                      how = 'left', on = ['tweet_id'])

df_twitter.info()
```
